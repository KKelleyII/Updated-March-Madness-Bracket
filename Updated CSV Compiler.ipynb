```python
import csv
import re

# Define the file paths
input_file = r"C:\Users\kevin\OneDrive\Desktop\March Madness Model\Mar 9 GTD.csv"
output_file = r"C:\Users\kevin\OneDrive\Desktop\March Madness Model\Mar 9 GTDCleaned.csv"

# Regex to split "Team Name    93   extra stuff"
field_pattern = re.compile(r'^(?P<team>.*?)\s*(?P<score>\d+)\s*(?P<extra>.*)$')

cleaned_data = []

with open(input_file, mode="r", newline="", encoding="utf-8-sig") as infile:
    reader = csv.reader(infile)

    for row_num, row in enumerate(reader, start=1):
        # Each row should look like:
        # [date, "Team 1    score", "Team 2    score   optional extra"]
        if len(row) != 3:
            print(f"Skipping row {row_num}: unexpected number of columns -> {row}")
            continue

        date, team1_field, team2_field = row

        m1 = field_pattern.match(team1_field.strip())
        m2 = field_pattern.match(team2_field.strip())

        if not m1 or not m2:
            print(f"Skipping row {row_num}: could not parse -> {row}")
            continue

        team1 = m1.group("team").strip()
        score1 = m1.group("score").strip()

        team2 = m2.group("team").strip()
        score2 = m2.group("score").strip()
        extra = m2.group("extra").strip()   # optional: OT count, neutral site, location, etc.

        cleaned_data.append([date, team1, team2, score1, score2, extra])

with open(output_file, mode="w", newline="", encoding="utf-8") as outfile:
    writer = csv.writer(outfile)
    writer.writerow(["Date", "Team 1", "Team 2", "Score 1", "Score 2", "Extra"])
    writer.writerows(cleaned_data)

print(f"Cleaning complete! The data has been saved to '{output_file}'.")
```
